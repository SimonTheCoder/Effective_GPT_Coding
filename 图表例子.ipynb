2012年，我国销售汽车1931万辆
2013年，我国销售汽车2198万辆

2015年，我国销售汽车2460万辆

```python
import matplotlib.pyplot as plt

years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
sales = [1931, 2198, 2349, 2460, 2803, 2888, 2808, 2577, 2531, 2627, 2686]

plt.plot(years, sales, marker='o')

plt.title("China Car Sales (2012-2022)")
plt.xlabel("Year")
plt.ylabel("Sales (millions)")
plt.xticks(years)
plt.grid(True)

plt.show()
```
